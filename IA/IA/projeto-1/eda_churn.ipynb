### Análise Bivariada

```python
# Hipótese: Clientes com contrato do tipo mensal são mais propensos ao cancelamento (Churn)
# Construir tabela de contingência entre Churn e Contract
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total')
```

```python
df_churn[['Churn', 'Contract']].value_counts().unstack().plot(kind='bar', title='Contagem de Churn por Contract')
```

```python
pd.crosstab(df_churn['Churn'], df_churn['Contract'], normalize=True, margins=True, margins_name='Total')
```

```python
pd.crosstab(df_churn['Churn'], df_churn['Contract'], normalize='index', margins=True, margins_name='Total')
```

Considerando o público total, os clientes de contrarto mensal representam 56%. Porém, se consdierarmos apenas os clientes que abandonaram o serviço 100% tinham o contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variáveis  categóricas (qualitativas)

# Executar um teste de hipótese chamado Chi-Square ou Teste do Qui-Quadrado de Pearson
# Em um teste de hipótese, duas hipóteses são formuladas:
# - Hipótese nula (H0): As duas variáveis são independentes.
# - Hipótese alternativa/complementar (H1): As duas variáveis não são idependentes.

# O teste serve para confirmar ou recusar a hipótese nula (H0).
# Quando a probabilidade de observarmos H0 é inferior a 5% (0.05), rejeitamos H0 e aceitamos H1.
```

```python
# Gerar um DF da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Cálcular os scores a P-values
%pip install scipy
from scipy.stats import chi2_contingency
```

```python
# Tupla = Lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
scores_churn_contract
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
pvalues_churn_contract
```

```python
# Apresentar números com decimais sem a notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar um DataFrame com os resultados
df_chi_scores_churn_contract = pd.DataFrame({
    'Qui2': scores_churn_contract,
    'P-Values': pvalues_churn_contract
})
df_chi_scores_churn_contract
```

Conforme o teste estatístico (chi-square), o P-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as não são independentes. Pelo Qui2 alto, podemos afirmar que há uma relação forte correlação.

```python
# Avaliar a correlação entre uma variável qualitativa e quantitativa
# Usar o mesmo teste de Chi-Square, mas trasnformar a variável quantitativa em qualitativa
# Hipótese: Cliente com menos de 6 meses de contrato é mais propenso ao cancelamento (Churn)

# Criar uma coluna nova no dataframe com base numa condição de outra coluna
# Coluna = 'TempoMenor6Meses'

import numpy as np
```

```python
df_churn['TempoMenor6Meses'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No')
df_churn.head(5)
```

```python
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Calcular os scores

chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_contract)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar um DataFrame com os resultados
df_chi_scores_churn_contract = pd.DataFrame({
    'Qui2': scores_churn_tenure,
    'P-Values': pvalues_churn_tenure
})

# Analisar os resultados
df_chi_scores_churn_contract
```

Como o p-value <= 0.05, rejeitamos a H0, ou seja, as variáveis não são independentes. E vale mencionar que com base do Qui2 resultante das análises, a correlação entre churn e tempo de contrato < 6 meses é menor do que a correlação entre churn e tipo de contrato

```python
# Correlações entre 2 variáveis númericas
# Tenure com TotalCharges
# A intuição é que quanto mais tempo de cintrato maior o valor pago

# Correlação entre 2 variáveis numéricas - Pearson
df_churn['tenure'].corr(df_churn['TotalCharges'])

```

```python
# Correlação entre 2 variáveis numéricas - Spearman
df_churn['tenure'].corr(df_churn['TotalCharges'], method='spearman')
```

Há uma forte correlação entre Tenure e TotalCharges, usando métodos estatísticos de correlação

```python
df_churn_month = df_churn[df_churn['Contract'] == 'Month-to-month']
df_churn_month
```

```python
# Histograma - checar visualmente se os dados seguem uma distribuição normal
df_churn_month['TotalCharges'].plot.hist(title='Histograma de TotalCharges (Contract: Month-to-month)')
```

```python
# Métodos de Tukey - IQR (Distribuição normal)
# IQR - Rabge Interquartile
# IQR = 3o quartil - 1o quartil
q1_TotalCharges_month = df_churn_month['TotalCharges'].quantile(0.25)
q3_TotalCharges_month = df_churn_month['TotalCharges'].quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites inferior e superior
limInf_TotalCharges_month = q1_TotalCharges_month - (1.5 * iqr_TotalCharges_month)
limSup_TotalCharges_month = q3_TotalCharges_month + (1.5 * iqr_TotalCharges_month)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month['TotalCharges'] < limInf_TotalCharges_month) |
               (df_churn_month['TotalCharges'] > limSup_TotalCharges_month)]
```

```python
# ZS (usando para distribuição normal)
# Indica qantos desvios padrão um ponto específico de dados  está distante da média
# zscore = (x- media) / desvio padrão

from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month['TotalCharges']))
z
```
